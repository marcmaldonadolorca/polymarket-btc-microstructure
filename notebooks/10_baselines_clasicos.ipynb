# 10 · Baselines clásicos de series temporales

**Qué pregunta responde.** ¿Cuánto rinde el instrumental clásico (suavizado exponencial, ARIMA, kNN-DTW) bajo el mismo protocolo económico que el especialista, y cuál es el listón que cualquier modelo profundo debe superar?

**Qué entradas usa.** `results/baselines_clasicos.csv` — acierto direccional y neto (ticks) de cada baseline sobre los casos congelados, con comisión taker real. No requiere el corpus privado.

**Qué produce.** La tabla comparativa de baselines (módulos 3 y 4 de la asignatura), que fija el nivel de referencia de la escalera del notebook 13.

**Conclusión (2 frases).** `naive-drift` y `ARIMA(1,1,1)` son los baselines más informativos (53–57 % de acierto direccional); Holt queda por detrás y kNN-DTW no supera a la deriva simple. Ninguno es rentable en neto taker bajo el régimen de comisiones de 2026 —lo esperado— pero establecen el listón que el aprendizaje profundo no logra batir (notebook 13).


```python
from pathlib import Path
import pandas as pd
ROOT = Path.cwd()
if not (ROOT / 'results').exists(): ROOT = ROOT.parent
bl = pd.read_csv(ROOT / 'results' / 'baselines_clasicos.csv')
bl.sort_values('acierto_pct', ascending=False).reset_index(drop=True)
```

## Lectura
El acierto direccional (`acierto_pct`) mide cuántas veces el signo predicho coincide con el signo del movimiento H60; el `neto_taker_ticks` incorpora la comisión oficial y sale negativo en todos los casos (el lado taker no es económico). La distinta `n` por método refleja que algunos (ARIMA, kNN-DTW) solo emiten señal en un subconjunto de casos.

Estos números son el nivel de referencia de la **escalera de modelos** (notebook 13): la vía profunda (MOMENT) no supera a `naive-drift` ni a `ARIMA`, y la curva de escalado confirma que más datos no cierran la brecha.
